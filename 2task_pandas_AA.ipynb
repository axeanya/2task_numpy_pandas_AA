**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
# way 1
mask_young = (data['age'] <35) & (data['age'] >=16)
mask_adult = (data['age'] <70) & (data['age'] >=35)
mask_retiree = (data['age'] <100) & (data['age'] >=70)

DS1 = data.copy()
DS1.loc[mask_young,'AgeGroup'] = 'Young'
DS1.loc[mask_adult,'AgeGroup'] = 'Adult'
DS1.loc[mask_retiree,'AgeGroup'] = 'Retiree'
# print(DS1.loc[:,['age', 'AgeGroup']])

# way 2
DS2 = data.copy()
DS2['AgeGroup'] = pd.cut(DS1['age'],
                         bins=[16, 35, 70, 100],
                         labels=['Young', 'Adult', 'Retiree'],
                         right=False)
print(DS2)
```
